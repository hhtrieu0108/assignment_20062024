```python
import pandas as pd
from sqlalchemy import create_engine
df = pd.read_csv("bank_transactions.csv")
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], format='%d/%m/%y').dt.date
```

```python
df['CustomerDOB'] = df['CustomerDOB'].apply(lambda x: pd.to_datetime(x,format='%d/%m/%Y').date() if isinstance(x,str) and len(x.split('/')[2]) == 4 else x )
```

```python
df['CustomerDOB'] = df['CustomerDOB'].apply(lambda x: pd.to_datetime(x,format='%d/%m/%y').date() if isinstance(x,str) and len(x.split('/')[2]) < 3 else x )
```
